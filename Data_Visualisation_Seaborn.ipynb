```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Seaborn has some datasets already and can be loaded easily.
sns.get_dataset_names()
```

```python
tips =sns.load_dataset('tips')
```

```python
#Univariate -one variable data visualisations
## Strip plot
sns.stripplot(tips.tip, jitter =True)
plt.ylabel ('Tip')
```

```python
#Grouping with stripplot
sns.stripplot(x='day',y='tip',data =tips, jitter =True)
```

```python
#Swarm plot
sns.swarmplot(x='tip',y='day', color= 'pink',data=tips,hue ="sex")
```

```python
# Swarm plot with different orientation
sns.swarmplot(x='tip',y='day',data =tips, hue='sex', orient='v')
```

```python
#Boxplot and Violin plot
sns.boxplot(x=tips.day, y=tips.tip)
plt.title('Boxplot',color='black', size =20)
```

```python
sns.violinplot(x=tips.day,y=tips.tip)
```

```python
#Combining plots
sns.violinplot(x=tips.day,y=tips.tip,color='grey',inner= None)
sns.stripplot(x=tips.day, y=tips.tip,jitter=True,size=4)
```

```python
#Subplots
plt.subplot(1,2,1)
sns.boxplot(x=tips.day, y=tips.tip)
plt.subplot(1,2,2)
sns.violinplot(x=tips.day,y=tips.tip)
```

```python
#Bivariate -2 variables data visualisation
#Joint plots
sns.jointplot(x='total_bill', y='tip',data=tips)
```

```python
#Density plot , KDE - Kernel density estimation
sns.jointplot(x='total_bill',y='tip',data=tips,kind= 'kde')
```

```python
#Pairplot
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```
